```python
import pandas as pd
AAPL = pd.read_csv('C:\\Users\\Arunima Chaudhary\\Downloads\\indicator CDIAC carbon_dioxide_total_emissions.csv')
```

```python
df2=AAPL.set_index("CO2 emission total")
```

```python
India = df2.loc["India","1990":"2011"]
China = df2.loc["China","1990":"2011"]
USA = df2.loc["United States","1990":"2011"]
Russia = df2.loc["Russia","1990":"2011"]
Japan = df2.loc["Japan","1990":"2011"]
Germany = df2.loc["Germany","1990":"2011"]
Iran = df2.loc["Iran","1990":"2011"]
SKorea = df2.loc["South Korea","1990":"2011"]
Canada = df2.loc["Canada","1990":"2011"]
SArabia = df2.loc["Saudi Arabia","1990":"2011"]
```

```python
Ch = pd.DataFrame({'year':China.index, 'emission':China.values})
In = pd.DataFrame({'year':India.index, 'emission':India.values})
US = pd.DataFrame({'year':USA.index, 'emission':USA.values})
Ru = pd.DataFrame({'year':Russia.index, 'emission':Russia.values})
Ja = pd.DataFrame({'year':Japan.index, 'emission':Japan.values})
Ge = pd.DataFrame({'year':Germany.index, 'emission':Germany.values})
Ir = pd.DataFrame({'year':Iran.index, 'emission':Iran.values})
SK = pd.DataFrame({'year':SKorea.index, 'emission':SKorea.values})
Ca = pd.DataFrame({'year':Canada.index, 'emission':Canada.values})
SA = pd.DataFrame({'year':SArabia.index, 'emission':SArabia.values})
```

```python
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool
from bokeh.models import NumeralTickFormatter
```

```python
ds = ColumnDataSource(data=dict(x=In[In.columns[1]], 
                                y1=Ch[Ch.columns[0]], 
                                y2=US[US.columns[0]], 
                                y3=Ru[Ru.columns[0]],
                                y4=Ru[Ru.columns[0]],
                                y5=Ru[Ru.columns[0]],
                                y6=Ru[Ru.columns[0]],
                                y7=Ru[Ru.columns[0]],
                                y8=Ru[Ru.columns[0]],
                                y9=Ru[Ru.columns[0]],
                                y10=Ru[Ru.columns[0]],
                               ))
max( ds.data['x'] )
```
